```
import numpy as np

import matplotlib.pyplot as plot

#create x and y axises
sinX = np.arange(0,6.4, 0.1);

sinY = np.sin(sinX)

plot.plot(sinX, sinY)

 
# Give a title for the sine wave plot
plot.title('Sine graph')
plot.xlabel('X')
plot.ylabel('Y')

plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
```

```
#create x and y axises
cosX = np.arange(0,6.4, 0.1)

cosY = np.cos(cosX)

plot.plot(cosX, cosY)

 
# Give a title for the sine wave plot
plot.title('cosine graph')
plot.xlabel('X')
plot.ylabel('Y')

plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
```
